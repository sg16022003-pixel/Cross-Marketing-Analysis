```python
# Historical cryptocurrencies prices insertion from df_bitcoin_prices
import sqlite3
import pandas as pd
from sqlalchemy import create_engine, text
from pycoingecko import CoinGeckoAPI # Import CoinGeckoAPI

# Create an SQLAlchemy engine
engine = create_engine('sqlite:///market_analytics.db')

# Initialize the API client
cg = CoinGeckoAPI()

# Define the coin_id
coin_id = 'bitcoin'

# Fetch historical market data (365 days) using pycoingecko
# 'days=365' automatically gives daily data (00:00 UTC)
data = cg.get_coin_market_chart_by_id(id=coin_id, vs_currency='usd', days=365)

# Extract the 'prices' list
# Data format is: [[timestamp, price], [timestamp, price], ...]
prices = data['prices']

# Create a DataFrame
df_bitcoin_prices = pd.DataFrame(prices, columns=['timestamp', 'price_usd'])

# Convert timestamp to datetime and then to 'YYYY-MM-DD' date string
df_bitcoin_prices['date'] = pd.to_datetime(df_bitcoin_prices['timestamp'], unit='ms').dt.strftime('%Y-%m-%d')

# Add coin_id column
df_bitcoin_prices['coin_id'] = coin_id

# Select and reorder columns to match the desired schema for Crypto_prices table
df_bitcoin_prices = df_bitcoin_prices[['coin_id', 'date', 'price_usd']]

# Explicitly drop the table if it exists to avoid reflection issues with 'replace'
try:
    with engine.connect() as connection:
        connection.execute(text("DROP TABLE IF EXISTS Crypto_prices"))
        connection.commit()
    print("Table 'Crypto_prices' dropped if it existed.")
except Exception as e:
    print(f"Error dropping table: {e}")

# Insert historical cryptocurrencies prices into 'Crypto_prices' table.
df_bitcoin_prices.to_sql('Crypto_prices', engine, if_exists='replace', index=False)

print("Historical cryptocurrencies prices inserted into 'Crypto_prices' table.")
```

```python
import sqlite3
import pandas as pd
import requests

# 1. Connect to SQLite (creates 'stock.db' if it doesn't exist)
conn = sqlite3.connect('stock.db')
cursor = conn.cursor()

print("Connected to stock.db")
```

```python
import yfinance as yf
import pandas as pd
import sqlite3
from sqlalchemy import create_engine

# Define the tickers you want to download
tickers = ['^GSPC', '^IXIC', '^NSEI']
start_date = "2020-01-01"
end_date = "2026-01-12"

# Download data for all tickers without group_by. This often returns a (Metric, Ticker) MultiIndex
stocks_df_raw = yf.download(tickers, start=start_date, end=end_date)

# Flatten the MultiIndex columns to a single level, e.g., 'Open_^GSPC'
# This makes column access much simpler.
stocks_df_raw.columns = ['_'.join(col).strip() for col in stocks_df_raw.columns.values]

# Create a database connection
engine = create_engine('sqlite:///stocks.db')
conn = engine.connect()

all_stocks_data_for_db = []

# Iterate through each ticker to process its data
for ticker_symbol in tickers:
    # Select columns relevant to the current ticker
    # Look for 'Open_^GSPC', 'High_^GSPC', etc.
    cols_for_ticker = [col for col in stocks_df_raw.columns if col.endswith(f'_{ticker_symbol}')]
    df_single_ticker = stocks_df_raw[cols_for_ticker].copy()

    # Rename columns to standard names ('open', 'high', etc.)
    # Example: 'Open_^GSPC' -> 'open'
    new_column_names = {col: col.split('_')[0].lower() for col in df_single_ticker.columns}
    df_single_ticker = df_single_ticker.rename(columns=new_column_names)

    # Reset index to make 'Date' a column
    df_single_ticker = df_single_ticker.reset_index()
    df_single_ticker = df_single_ticker.rename(columns={'Date': 'date'})

    # Add the ticker column
    df_single_ticker['ticker'] = ticker_symbol

    # Handle 'Adj Close': if 'close' is not present but 'adj close' is, rename 'adj close' to 'close'.
    # Otherwise, if 'adj close' is present, drop it since 'close' is likely already adjusted.
    if 'adj close' in df_single_ticker.columns and 'close' not in df_single_ticker.columns:
        df_single_ticker = df_single_ticker.rename(columns={'adj close': 'close'})
    elif 'adj close' in df_single_ticker.columns:
        df_single_ticker = df_single_ticker.drop(columns=['adj close'])

    # Select only the columns needed for the database schema
    all_stocks_data_for_db.append(df_single_ticker[['date', 'ticker', 'open', 'high', 'low', 'close', 'volume']])

# Concatenate all individual ticker DataFrames into a single DataFrame
stocks_final_df = pd.concat(all_stocks_data_for_db, ignore_index=True)

# Ensure 'date' column is in a format compatible with SQLite DATE type (YYYY-MM-DD)
stocks_final_df['date'] = pd.to_datetime(stocks_final_df['date']).dt.strftime('%Y-%m-%d')

# Insert the stock price data into the Stock_prices table
# Using 'replace' for if_exists to ensure fresh data each run if table exists
stocks_final_df.to_sql('Stock_prices', engine, if_exists='replace', index=False)

print("Stock price data inserted into 'Stock_prices' table.")

# Now, try to query the table again
query = """
SELECT
    date,
    open,
    high,
    low,
    close,
    volume,
    ticker
FROM
    Stock_prices
ORDER BY
    date ASC
LIMIT 10;
"""
stocks_df_from_db = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the results
print("Top 10 rows from Stock_prices table after insertion:")
print(stocks_df_from_db)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('market_analytics.db')

query="""
SELECT
    symbol,
    name,
    circulating_supply,
    total_supply,
    (CAST(circulating_supply AS REAL) / total_supply) * 100 AS supply_percent
FROM
    Cryptocurrencies
WHERE
    total_supply > 0
    AND (CAST(circulating_supply AS REAL) / total_supply) >= 0.90
ORDER BY
    supply_percent DESC
LIMIT 3;
"""
top_3_crypto = pd.read_sql_query(query, conn)
conn.close()

print("Top 3 Cryptocurrencies by Circulating Supply Percentage:")
print(top_3_crypto)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('market_analytics.db')

query="""
SELECT
    C.Name,
    C.symbol,
    CP.current_price,
    C.ath,
    ((CP.current_price / C.ath) * 100) AS percent_of_ath
FROM
    crypto_prices AS CP
JOIN
    Cryptocurrencies AS C ON CP.ID = C.ID
WHERE
    C.ath > 0
    AND (CP.current_price / C.ath) >= 0.90
ORDER BY
    percent_of_ath DESC
LIMIT 3;
"""
top_3_crypto = pd.read_sql_query(query, conn)
conn.close()

print("Top 3 Cryptocurrencies by Percentage of All-Time High:")
print(top_3_crypto)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('market_analytics.db')

query = """
WITH price_data AS (
    SELECT
        (SELECT current_price FROM Crypto_prices WHERE date = '2024-09-01' AND ID = 'bitcoin') AS start_price,
        (SELECT current_price FROM Crypto_prices WHERE date = '2025-09-01' AND ID = 'bitcoin') AS end_price
)
SELECT
    start_price,
    end_price,
    ((end_price - start_price) / start_price) * 100 AS percent_change
FROM price_data;
"""

price_change_df = pd.read_sql_query(query, conn)
conn.close()

print("Bitcoin Price Change (2024-09-01 to 2025-09-01):")
print(price_change_df)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('market_analytics.db')

query="""
SELECT
    MAX(price_usd) AS highest_oil_price,
    Date
FROM
    Oil_prices
WHERE
    Date >= DATE('now', '-5 years')
GROUP BY
    Date
ORDER BY
    highest_oil_price DESC
LIMIT 1;
"""

highest_oil_price_df = pd.read_sql_query(query, conn)
conn.close()

print("Highest oil price in the last 5 years:")
print(highest_oil_price_df)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('market_analytics.db')

query = """
SELECT
    CP.date AS price_date,
    CP.price_usd AS btc_price,
    OP.price_usd AS oil_price
FROM
    Crypto_prices AS CP
JOIN
    Oil_prices AS OP ON CP.date = STRFTIME('%Y-%m-%d', OP.Date)
WHERE
    CP.coin_id = 'bitcoin'
    AND CP.date BETWEEN '2025-01-01' AND '2025-12-31'
ORDER BY
    CP.date ASC;
"""

joined_prices_df = pd.read_sql_query(query, conn)
conn.close()

print("Bitcoin and Oil Prices for 2025:")
print(joined_prices_df.head())
```

```python
import sqlite3
import pandas as pd

# --- Step 1: Fetch Nifty 50 data from stocks.db ---
conn_stocks = sqlite3.connect('stocks.db')
query_nifty = """
SELECT
    date AS trade_date,
    open AS open_price,
    close AS close_price
FROM
    Stock_prices
WHERE
    ticker = '^NSEI'
    AND date BETWEEN '2025-01-01' AND '2026-02-24'
ORDER BY
    date DESC;
"""
df_nifty = pd.read_sql_query(query_nifty, conn_stocks)
conn_stocks.close()

# --- Step 2: Fetch Cryptocurrency prices from market_analytics.db ---
conn_crypto = sqlite3.connect('market_analytics.db')
query_crypto = """
SELECT
    date AS trade_date,
    coin_id,
    price_usd
FROM
    Crypto_prices
WHERE
    coin_id IN ('bitcoin', 'ethereum', 'solana')
    AND date BETWEEN '2025-01-01' AND '2026-02-24'
ORDER BY
    date DESC;
"""
df_crypto = pd.read_sql_query(query_crypto, conn_crypto)
conn_crypto.close()

# Pivot crypto data to have separate columns for each coin's price
df_crypto_pivot = df_crypto.pivot_table(index='trade_date', columns='coin_id', values='price_usd').reset_index()
df_crypto_pivot.columns.name = None # Remove the columns name 'coin_id'

# --- Step 3: Merge DataFrames ---
merged_df = pd.merge(df_nifty, df_crypto_pivot, on='trade_date', how='inner')

# --- Step 4: Perform calculations ---
merged_df['nifty_pct_change'] = round(((merged_df['close_price'] - merged_df['open_price']) / merged_df['open_price']) * 100, 2)

# Rename columns to match the original query's aliases for crypto prices
merged_df = merged_df.rename(columns={
    'bitcoin': 'btc_price',
    'ethereum': 'eth_price',
    'solana': 'sol_price'
})

# --- Step 5: Display results ---
print("Combined Daily Data (Nifty 50, Bitcoin, Ethereum, Solana):")
# Adjusting print statement to only display columns that are currently available
# The KeyError happened because 'eth_price' and 'sol_price' were not in merged_df
# as the Crypto_prices table only contained Bitcoin data.
print(merged_df[['trade_date', 'close_price', 'btc_price', 'nifty_pct_change']].head())
```

```python
import sqlite3
import pandas as pd

# --- Step 1: Fetch S&P 500 data from stocks.db ---
conn_stocks = sqlite3.connect('stocks.db')
query_spx = """
SELECT
    date AS trade_date,
    close
FROM
    Stock_prices
WHERE
    ticker = '^GSPC'
    AND date BETWEEN '2025-01-01' AND '2026-02-25'
ORDER BY
    date ASC;
"""
df_spx = pd.read_sql_query(query_spx, conn_stocks)
conn_stocks.close()

# --- Step 2: Fetch Oil prices from market_analytics.db ---
conn_oil = sqlite3.connect('market_analytics.db')
query_oil = """
SELECT
    Date AS trade_date,
    price_usd
FROM
    Oil_prices
WHERE
    trade_date BETWEEN '2025-01-01' AND '2026-02-25'
ORDER BY
    trade_date ASC;
"""
df_oil = pd.read_sql_query(query_oil, conn_oil)
conn_oil.close()

# Ensure 'trade_date' columns are in datetime format for merging and calculations
df_spx['trade_date'] = pd.to_datetime(df_spx['trade_date'])
df_oil['trade_date'] = pd.to_datetime(df_oil['trade_date'])

# --- Step 3: Calculate daily percentage change for S&P 500 ---
df_spx['prev_close'] = df_spx['close'].shift(1)
df_spx['sp500_daily_pct'] = round(((df_spx['close'] - df_spx['prev_close']) / df_spx['prev_close']) * 100, 2)

# --- Step 4: Calculate daily percentage change for Oil prices ---
df_oil['prev_price'] = df_oil['price_usd'].shift(1)
df_oil['oil_daily_pct'] = round(((df_oil['price_usd'] - df_oil['prev_price']) / df_oil['prev_price']) * 100, 2)

# --- Step 5: Merge the two DataFrames ---
merged_df = pd.merge(df_spx[['trade_date', 'close', 'sp500_daily_pct']],
                     df_oil[['trade_date', 'price_usd', 'oil_daily_pct']],
                     on='trade_date', how='inner')

# Rename columns for clarity as in the original SQL query
merged_df = merged_df.rename(columns={
    'close': 'sp500_level',
    'price_usd': 'oil_price_wti'
})

# --- Step 6: Display results ---
print("Combined Daily Data (S&P 500 and WTI Oil, including daily percentage changes):")
print(merged_df.head())
```

```python
import sqlite3
import pandas as pd
from pycoingecko import CoinGeckoAPI
from datetime import datetime, timedelta

# Initialize CoinGeckoAPI
cg = CoinGeckoAPI()

# Define a consistent date range for the last 365 days
current_date = datetime.now()
from_timestamp = int((current_date - timedelta(days=365)).timestamp())
to_timestamp = int(current_date.timestamp())
from_date_str = (current_date - timedelta(days=365)).strftime('%Y-%m-%d')
to_date_str = current_date.strftime('%Y-%m-%d')

# --- 1. Fetch S&P 500 data from stocks.db ---
conn_stocks = sqlite3.connect('stocks.db')
query_spx = f"""
SELECT
    date AS trade_date,
    close AS sp500_close
FROM
    Stock_prices
WHERE
    ticker = '^GSPC'
    AND date BETWEEN '{from_date_str}' AND '{to_date_str}'
ORDER BY
    date ASC;
"""
df_spx = pd.read_sql_query(query_spx, conn_stocks)
conn_stocks.close()
df_spx['trade_date'] = pd.to_datetime(df_spx['trade_date'])
df_spx = df_spx.sort_values(by='trade_date')
df_spx['prev_sp'] = df_spx['sp500_close'].shift(1)
df_spx['sp500_daily_pct'] = round(((df_spx['sp500_close'] - df_spx['prev_sp']) / df_spx['prev_sp']) * 100, 2)

# --- 2. Fetch Nifty 50 data from stocks.db ---
conn_stocks = sqlite3.connect('stocks.db') # Re-establish connection or ensure it's open
query_nifty = f"""
SELECT
    date AS trade_date,
    close AS nifty_close
FROM
    Stock_prices
WHERE
    ticker = '^NSEI'
    AND date BETWEEN '{from_date_str}' AND '{to_date_str}'
ORDER BY
    date ASC;
"""
df_nifty = pd.read_sql_query(query_nifty, conn_stocks)
conn_stocks.close()
df_nifty['trade_date'] = pd.to_datetime(df_nifty['trade_date'])
df_nifty = df_nifty.sort_values(by='trade_date')

# --- 3. Fetch Bitcoin data from market_analytics.db ---
conn_crypto = sqlite3.connect('market_analytics.db')
query_btc = f"""
SELECT
    date AS trade_date,
    price_usd AS btc_price_usd
FROM
    Crypto_prices
WHERE
    coin_id = 'bitcoin'
    AND date BETWEEN '{from_date_str}' AND '{to_date_str}'
ORDER BY
    date ASC;
"""
df_btc = pd.read_sql_query(query_btc, conn_crypto)
conn_crypto.close()
df_btc['trade_date'] = pd.to_datetime(df_btc['trade_date'])
df_btc = df_btc.sort_values(by='trade_date')
df_btc['prev_btc'] = df_btc['btc_price_usd'].shift(1)
df_btc['btc_daily_pct'] = round(((df_btc['btc_price_usd'] - df_btc['prev_btc']) / df_btc['prev_btc']) * 100, 2)

# --- 4. Fetch Ethereum data from CoinGecko API ---
etherum_data = cg.get_coin_market_chart_range_by_id(
    id='ethereum', vs_currency='usd', from_timestamp=from_timestamp, to_timestamp=to_timestamp
)
df_eth = pd.DataFrame(etherum_data['prices'], columns=['timestamp', 'eth_price_usd'])
df_eth['trade_date'] = pd.to_datetime(df_eth['timestamp'], unit='ms').dt.date
df_eth['trade_date'] = pd.to_datetime(df_eth['trade_date'])
df_eth = df_eth.sort_values(by='trade_date')

# --- 5. Fetch Tether data from CoinGecko API ---
tether_data = cg.get_coin_market_chart_range_by_id(
    id='tether', vs_currency='usd', from_timestamp=from_timestamp, to_timestamp=to_timestamp
)
df_tether = pd.DataFrame(tether_data['prices'], columns=['timestamp', 'tether_price_usd'])
df_tether['trade_date'] = pd.to_datetime(df_tether['timestamp'], unit='ms').dt.date
df_tether['trade_date'] = pd.to_datetime(df_tether['trade_date'])
df_tether = df_tether.sort_values(by='trade_date')
df_tether['usdt_peg_ratio'] = round(df_tether['tether_price_usd'] / 1.0, 4) # Tether should be pegged to 1 USD

# --- 6. Merge all DataFrames ---
merged_df = pd.merge(df_spx[['trade_date', 'sp500_close', 'sp500_daily_pct']], df_nifty[['trade_date', 'nifty_close']], on='trade_date', how='inner')
merged_df = pd.merge(merged_df, df_btc[['trade_date', 'btc_price_usd', 'btc_daily_pct']], on='trade_date', how='inner')
merged_df = pd.merge(merged_df, df_eth[['trade_date', 'eth_price_usd']], on='trade_date', how='inner')
merged_df = pd.merge(merged_df, df_tether[['trade_date', 'usdt_peg_ratio']], on='trade_date', how='inner')

# --- 7. Display results ---
print("Combined Market Data (Stocks and Cryptos with Daily % Changes):")
print(merged_df.head())
```

```python
import sqlite3
import pandas as pd

# --- Step 1: Fetch S&P 500 data from stocks.db ---
conn_stocks = sqlite3.connect('stocks.db')
query_spx = """
SELECT
    date AS trade_date,
    close AS sp500_close
FROM
    Stock_prices
WHERE
    ticker = '^GSPC'
    AND date BETWEEN '2025-01-01' AND '2026-02-25'
ORDER BY
    date ASC;
"""
df_spx = pd.read_sql_query(query_spx, conn_stocks)
conn_stocks.close()

# --- Step 2: Fetch Oil prices from market_analytics.db ---
conn_analytics = sqlite3.connect('market_analytics.db')
query_oil = """
SELECT
    Date AS trade_date,
    price_usd AS oil_wti_price
FROM
    Oil_prices
WHERE
    Date BETWEEN '2025-01-01' AND '2026-02-25'
ORDER BY
    Date ASC;
"""
df_oil = pd.read_sql_query(query_oil, conn_analytics)

# --- Step 3: Fetch Bitcoin data from market_analytics.db ---
query_btc = """
SELECT
    date AS trade_date,
    price_usd AS btc_price
FROM
    Crypto_prices
WHERE
    coin_id = 'bitcoin'
    AND date BETWEEN '2025-01-01' AND '2026-02-25'
ORDER BY
    date ASC;
"""
df_btc = pd.read_sql_query(query_btc, conn_analytics)
conn_analytics.close()

# Ensure 'trade_date' columns are in datetime format for merging and calculations
df_spx['trade_date'] = pd.to_datetime(df_spx['trade_date'])
df_oil['trade_date'] = pd.to_datetime(df_oil['trade_date'])
df_btc['trade_date'] = pd.to_datetime(df_btc['trade_date'])

# --- Step 4: Calculate daily percentage changes ---
df_spx['prev_close'] = df_spx['sp500_close'].shift(1)
df_spx['sp500_return'] = round(((df_spx['sp500_close'] - df_spx['prev_close']) / df_spx['prev_close']) * 100, 2)

df_oil['prev_price'] = df_oil['oil_wti_price'].shift(1)
df_oil['oil_return'] = round(((df_oil['oil_wti_price'] - df_oil['prev_price']) / df_oil['prev_price']) * 100, 2)

df_btc['prev_price'] = df_btc['btc_price'].shift(1)
df_btc['btc_return'] = round(((df_btc['btc_price'] - df_btc['prev_price']) / df_btc['prev_price']) * 100, 2)

# --- Step 5: Merge all DataFrames ---
merged_df = pd.merge(df_spx[['trade_date', 'sp500_close', 'sp500_return']],
                     df_oil[['trade_date', 'oil_wti_price', 'oil_return']],
                     on='trade_date', how='inner')
merged_df = pd.merge(merged_df,
                     df_btc[['trade_date', 'btc_price', 'btc_return']],
                     on='trade_date', how='inner')

# --- Step 6: Display results ---
print("Combined Daily Data (S&P 500, WTI Oil, and Bitcoin with Daily Returns):")
print(merged_df.head())
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import sqlite3
import yfinance as yf
from datetime import datetime, timedelta

# --- DATABASE & DATA FETCHING ---
@st.cache_resource
def init_db():
    conn = sqlite3.connect('market_data.db', check_same_thread=False)
    # Generate dummy data for Dashboard/SQL Runner
    dates = [(datetime.now() - timedelta(days=i)).strftime('%Y-%m-%d') for i in range(100)]
    dates.reverse()

    btc = pd.DataFrame({'date': dates, 'price': [90000 + (i*150) for i in range(100)]})
    oil = pd.DataFrame({'date': dates, 'price': [70 + (i*0.2) for i in range(100)]})
    sp500 = pd.DataFrame({'date': dates, 'price': [5000 + (i*5) for i in range(100)]})
    nifty = pd.DataFrame({'date': dates, 'price': [22000 + (i*10) for i in range(100)]})

    btc.to_sql('bitcoin', conn, if_exists='replace', index=False)
    oil.to_sql('oil', conn, if_exists='replace', index=False)
    sp500.to_sql('sp500', conn, if_exists='replace', index=False)
    nifty.to_sql('nifty', conn, if_exists='replace', index=False)
    return conn

conn = init_db()

# --- SIDEBAR NAVIGATION ---
st.sidebar.title("Navigation")
page = st.sidebar.radio("Go to", ["Market Dashboard", "SQL Query Runner", "Top 3 Crypto Analysis"])

# --- PAGE 1 & 2 (Previous Logic) ---
if page == "Market Dashboard":
    st.title(" Market Dashboard")
    col1, col2 = st.columns(2)
    start_date = col1.date_input("Start Date", datetime.now() - timedelta(days=30))
    end_date = col2.date_input("End Date", datetime.now())

    query = f"SELECT b.date, b.price AS BTC, o.price AS Oil, s.price AS SP500, n.price AS NIFTY FROM bitcoin b JOIN oil o ON b.date = o.date JOIN sp500 s ON b.date = s.date JOIN nifty n ON b.date = n.date WHERE b.date BETWEEN '{start_date}' AND '{end_date}'"
    df = pd.read_sql(query, conn)
    if not df.empty:
        st.dataframe(df, use_container_width=True)

elif page == "SQL Query Runner":
    st.title(" SQL Query Runner")
    queries = {"Join Summary": "SELECT b.date, b.price as BTC, o.price as Oil FROM bitcoin b JOIN oil o ON b.date = o.date LIMIT 10"}
    sql_input = st.text_area("SQL Editor", queries["Join Summary"], height=150)
    if st.button("Run Query"):
        st.dataframe(pd.read_sql(sql_input, conn), use_container_width=True)

# --- PAGE 3: TOP 3 CRYPTO ANALYSIS ---
else:
    st.title(" Top 3 Crypto Analysis")

    # 1. Selection
    crypto_mapping = {
        "Bitcoin (BTC)": "BTC-USD",
        "Ethereum (ETH)": "ETH-USD",
        "Tether (USDT)": "USDT-USD"
    }
    selected_name = st.selectbox("Select Cryptocurrency", list(crypto_mapping.keys()))
    ticker = crypto_mapping[selected_name]

    # 2. Date Filter
    col1, col2 = st.columns(2)
    start_dt = col1.date_input("From", datetime.now() - timedelta(days=60))
    end_dt = col2.date_input("To", datetime.now())

    # 3. Fetch Real Data
    with st.spinner(f'Fetching {selected_name} data...'):
        data = yf.download(ticker, start=start_dt, end=end_dt)

    if not data.empty:
        # Flatten columns if multi-index (common in newer yfinance)
        data = data[['Close']].reset_index()
        data.columns = ['Date', 'Price']

        # 4. Display Trend Chart
        st.subheader(f"{selected_name} Price Trend")
        st.line_chart(data.set_index('Date'))

        # 5. Display Table
        st.subheader("Daily Price Data")
        st.dataframe(data.sort_values(by='Date', ascending=False), use_container_width=True)
    else:
        st.error("No data found for this period.")
```
